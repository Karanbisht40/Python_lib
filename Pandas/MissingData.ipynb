Missing Data

Finding missing Number

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isnull()   #find is any nan value
```

```python
df.isnull().sum()
```

```python
df.isna().any()
```

```python
df[df.isnull().any(axis=1)]
```

Removing missing number

```python
df
```

```python
df.dropna(
)
```

```python
df.dropna(thresh=3) #Keep rows with at least 3 non-missing values
```

```python
df.dropna(axis=1)
```

FIlling missing number

```python
df
```

```python
df.fillna(5)
```

```python
#filling woth different values
values= {'A':0,'B':100, 'C':900,'D':300}
df.fillna(value=values)
```

```python
#filling with means
df.fillna(df.mean())
```
