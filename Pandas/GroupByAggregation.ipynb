groupby aggregation

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#group by category
data = df.groupby('Category')
data
for i, v in data:
    print(i)
    print(v)
```

```python
#calculate the sum
data = df.groupby('Category')['Sales'].sum()
data
```

```python
#group by sales and calculate the sum
data = df.groupby('Store')['Sales'].sum()
data
# for i,v in data:
#     print(i)
#     print(v)
```

```python
#group by multiple column
data = df.groupby(['Category','Store'])['Sales'].sum()
data
```
