Pandas series

```python
import numpy as np
import pandas as pd
```

```python
labels =['a','b','c']
lists = [10,20,30,]
arr= np.array([10,20,30])
d = {1:10,2:20,3:30}
```

```python
pd.Series(lists)
```

```python
pd.Series(lists, index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
