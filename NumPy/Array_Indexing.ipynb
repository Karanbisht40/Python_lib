Array indexing and Slicing of vectors

```python
import numpy as np
```

```python
arr = np.arange(11,21)
arr
```

```python
#indexing
arr[1]  #+ve
```

```python
#slicing
arr[2:6]
```

```python
arr[:6]
```

```python
arr[:]
```

```python
arr[3:]
```

```python
#gap
arr[3:9:2]
```

```python
#reverse
print(arr[::-1])
```

indexing and slicing of matrix


```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
print(arr[1]) #particular row
```

```python
print(arr[0,4]) #paticular idx
```

```python
#slicing
arr[0:2,1:3 ]
```

```python
arr[3:,3:]
```

```python
#first two row
arr[0:2, :]
```

```python
#last two rows
arr[4:,:]
```

```python
#col
arr[:,2]
```

```python
#first two col
arr[:, 0:2]
```

```python
 #last two col
arr[:,3:]
```

```python
#reverse the entrie matrix
print(arr[::-1, ::-1])
```

```python
arr[0:2, 2:4]
```

```python
arr[4:,3:]
```
