Deep and Shallow copy

```python
#shallow copy
arr = np.array([10,20,30,40]) 
```

```python
view = arr.view()
print(view)
```

```python
view[0] = 100
print(view)
print(arr)
```

```python
#deep copy
arr = np.array([10,20,30,40]) 
```

```python
copy = arr.copy()
print(copy)
```

```python
copy[0]= 100
print(copy)
print(arr)
```
