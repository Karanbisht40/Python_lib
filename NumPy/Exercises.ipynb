```python
s = np.array([
    [5, 3, 4, 6, 7, 8, 9, 1, 2],
    [6, 7, 2, 1, 9, 5, 3, 4, 8],
    [1, 9, 8, 3, 4, 2, 5, 6, 7],

    [8, 5, 9, 7, 6, 1, 4, 2, 3],
    [4, 2, 6, 8, 5, 3, 7, 9, 1],
    [7, 1, 3, 9, 2, 4, 8, 5, 6],
    
    [9, 6, 1, 5, 3, 7, 2, 8, 4],
    [2, 8, 7, 4, 1, 9, 6, 3, 5],
    [3, 4, 5, 2, 8, 6, 1, 7, 9]
])
#sudoku
```

```python
#for rows
b =np.sum(s, axis=1)
for i in b:
    if i != 45:
        print("sudoko is not valid")
        break
else:
    print("sudoku is valid")
```

```python
#for columns
col=np.sum(s, axis=0)
for i in col:
    if i != 45:
        print("sudoko is not valid")
        break
else:
    print("sudoku is valid")
```

```python
#Get the shape of the matrix.
data.shape
```

```python
#Find the average age of students.
data[:,0:1].mean()
```

```python
#Extract Math marks of all students.
data[:, 1:2]
```

```python
#Find the highest Science mark.
data[:,2:].max()
```

```python
#Get details of the student who scored more than 90 in Math.
data[data[:,1]>90]
```

import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])

```python
#Find how many students are younger than 19.
len(data[data[:,0] <19])
```

```python
#Calculate the average marks in each subject (column-wise mean).
np.mean(data[:,1:],axis=0)
```

```python
#Get data of students who scored at least 80 in both subjects.
data[(data[:,1] >=80) & (data[:,2] >=80)]
```

```python
#Replace all Science marks < 75 with 0.
data[:,2][data[:,2] <=75]= 0
data
```
