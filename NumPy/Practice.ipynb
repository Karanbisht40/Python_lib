```python
import numpy as np
```

```python
x = np.array([1,2,3])
y= np.array([2,4,5])
```

```python
print(x+y)
print(x-y)
```

```python
#absolute value
a= np.array([-3,3,4,-5,-5,-1])
print(np.absolute(a))
```

```python
#binary operator
a= np.array([1,2,3,4])
b= np.array([9,2,2,4])
res = np.add(a,b)
print(res)
```
